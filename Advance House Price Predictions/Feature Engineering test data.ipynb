### Missing Values

```python
#Handling categorical features with missing values
categorical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=="O"]
for feature in categorical_nan:
    print("{} has {} %  missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
#Replacing missing valueswith the new label
def cat_fillna(dataset,categorical_nan):
    data=df.copy()
    data[categorical_nan]=data[categorical_nan].fillna("Missing")
    return data
```

```python
df=cat_fillna(df,categorical_nan)
df[categorical_nan].isnull().sum()
```

```python
df.head()
```

```python
#Handling numerical data with missing values
numerical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!="O"]
for feature in numerical_nan:
    print("{} has {} % of missing values. ".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
#Replacing the numerical missing values

for feature in numerical_nan:
    #We will replace using median since there are lotof outliers
    median_values=df[feature].median()
    
    #create a feature to capture nan values
    df[feature+"nan"]=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_values,inplace=True)
```

```python
df[numerical_nan].isnull().sum()
```

```python
#Temporal Variables(Date Time variables)
temp_features=[feature for feature in df.columns if "Yr" in feature or "Year" in feature ]
temp_features.remove('YrSold')
temp_features.remove('GarageYrBltnan')
temp_features
```

```python
for feature in temp_features:
    df[feature]=df['YrSold']-df[feature]
```
